```python
import pandas as pd
```

```python
df = pd.read_csv('../data/hpa.csv')
df
```

```python
def get_set(x):
    x = x[1:-1]
    x = x.replace("'", "")
    x = x.split(',')
    x = [i.strip() for i in x]
    return x

df['Locations'] = df.Locations.apply(lambda x: get_set(x))
```

```python
loc_dist = {}

for i in range(len(df)):
    locations = df.loc[i, 'Locations']
    rel = df.loc[i, 'Reliability']
    if rel != 'Enhanced':
        continue
    for location in locations:
        try:
            loc_dist[location] += 1
        except:
            loc_dist[location] = 1

dict(sorted(loc_dist.items(), key=lambda item: item[1]))
```
